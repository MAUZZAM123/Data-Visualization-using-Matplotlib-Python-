```python
import matplotlib.pyplot as plt
regions = ['East','West','North','South']
revenue = [1500,1000,3000,2000]

plt.pie(revenue,labels= regions,autopct='%1.1f%%', colors=['gold','silver','green','blue'])
plt.title('Revenue Contribution By region')
plt.show()
```

```python
import matplotlib.pyplot as plt
scores=[45, 38, 42, 50, 33, 29, 47, 41, 36, 39,44, 28, 35, 30, 43, 32, 26, 40, 31, 48,46, 37, 25, 34, 27, 49, 38, 29, 33, 42,40, 44, 36, 30, 39, 31, 45, 43, 32, 47,28, 35, 41, 37, 46]

plt.hist(scores,bins=10, color ='purple', edgecolor ='black')
plt.xlabel('Score range')
plt.ylabel('Number of Students')
plt.title('Score Distribution of students')
plt.show()
```

```python
hours_studies= [1,2,3,4,5,6,7,8]
exam_scores = [35,47,49,55,62,67,78,99]

plt.scatter(hours_studies,exam_scores,color='green', marker = 's',label ='Student Data')

plt.xlabel('Hours Study ')
plt.ylabel('Exam Score')
plt.title('Raltionship Between Study time and Exam score ')
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.scatter([1,2,3],[50,55,60],color ='green',marker ='^',label ='A')
plt.scatter([1,2,3],[40,45,55],color ='black',label ='B')

plt.xlabel('Hours Study ')
plt.ylabel('Exam Score')
plt.title('Comparison of two classes  ')
plt.legend()
plt.grid(True)
plt.show()
```

```python


x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar chart')

plt.show()
```

```python
fig , ax=plt.subplots(1,2,figsize=(10,5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y , color='Black')
ax[0].set_title('Line Plot')

ax[1].bar(x,y, color = 'gold')
ax[1].set_title('Bar chart')

fig.suptitle("Comparison of line and Bar chart")

plt.tight_layout()

plt.show()
```
